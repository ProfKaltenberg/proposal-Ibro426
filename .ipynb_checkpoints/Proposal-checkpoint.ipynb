# API Project Proposal: EV Charging Buildout and Electricity Prices

## Question  
Do higher electricity prices slow the growth of EV charging infrastructure across states?

## Background & Motivation  
This project looks at the relationship between electricity prices and the growth of EV charging infrastructure. Most discussions around electric vehicles focus on adoption, but the infrastructure side is just as important. Charging stations are costly to install and operate, especially fast chargers, and electricity is a direct operating cost. If electricity prices are high, it may reduce the incentive to expand charging networks, particularly in areas where profitability is already uncertain.

What makes this interesting is that not all states face the same electricity prices, and not all states are building out charging infrastructure at the same rate. Some states with high EV adoption still struggle with charger availability, while others have more stable growth. This suggests that cost conditions, not just demand, may play a role.

This project focuses on the supply side. Instead of asking where EVs are being adopted, it asks whether the cost of electricity itself affects how quickly charging infrastructure expands.

## APIs & Data  

### EIA API (Electricity Prices)  
This will be used to collect electricity prices by state over time. I will focus on the commercial electricity price, since public charging stations operate under commercial rate structures rather than residential ones.

### Open Charge Map API (Charging Stations)  
This will be used to collect EV charging station locations and metadata. It includes timestamps such as when a station was added to the database, which can be used to approximate growth over time.

### Census ACS API (Controls)  
This will be used to collect state-level controls such as population and median household income.

I have obtained API access for all three and will include screenshots at the end of the proposal.

## Variables  

### Variable 1: Chargers per 100,000 People  
This will be constructed using Open Charge Map data (charger counts) and Census population data. It will serve as the main outcome variable.

### Variable 2: Electricity Price (Commercial)  
This will come from the EIA API and will be measured in cents per kilowatt-hour. This is the main explanatory variable.

### Variable 3: Charger Price Penalty (Constructed Variable)  
Charger Price Penalty = (chargers per 100,000 people) / electricity price  

This is meant to capture how much charging infrastructure exists relative to how expensive electricity is. A lower value suggests that high electricity prices may be limiting infrastructure relative to population.

## Main Controls  

- Median household income  
- Population  
- Year fixed effects  

These help control for differences in economic conditions and scale across states.

## Design / Approach  

The project will construct a state-level panel dataset over time by combining electricity prices, charging station counts, and demographic controls.

The main idea is to examine whether states with higher electricity prices tend to have fewer chargers per capita or slower growth in chargers over time.

I may also extend the analysis by separating charger types (Level 2 vs DC fast chargers), since fast chargers are more sensitive to electricity costs and may respond differently.

## Why This Project Works  

This project is feasible because all data sources are accessible through APIs and can be merged at the state level. The variables are clearly defined, and the constructed variable adds a non-obvious dimension to the analysis.

It is also different from standard EV adoption projects because it focuses on infrastructure and cost rather than demand.
