# API Project Proposal: Minimum Wage Increases and County-Level Wages
## Question

How do state minimum wage increases affect county-level wages in low-wage industries, and are those effects stronger in places where the minimum wage is more binding?

## Background & Motivation

This project builds off my undergraduate thesis, where I looked at the effects of minimum wage increases on county-level wages across income groups. In that paper, I used county-industry wage data from the BEA and county-level controls from the ACS, and I focused on three industries that tend to have a large share of low-wage workers: Retail Trade, Accommodation and Food Services, and Other Services. The paper used a two-way fixed effects event study from 2014 to 2019 and compared counties in states that started raising minimum wages in 2016 to counties in states that did not make major changes during that period. It also split counties into high-income and low-income groups to see whether the effects looked different across local economic settings.

What makes this interesting to me now is that I do not want to just repeat the same paper. The original version had a clear weakness. The pre-treatment period was very short, and in the results the 2014 lead was statistically significant, which suggests that the treated and control counties may already have been moving differently before the policy change. That weakens the parallel trends assumption and makes the causal interpretation less clean than I would like.

So this project is basically a revised version of that idea. I want to keep the general topic, but improve the design by using a longer pre-treatment period, bringing in more states that adopted minimum wage increases at different times, and using a staggered treatment framework instead of treating 2016 like one single shock. I also want to add a new constructed variable, the Minimum Wage Bite Index, which should be a better way to capture how strongly the policy is actually expected to matter in a local labor market.

More broadly, this matters because minimum wage policy is usually studied at the state or national level, but the effects are not going to be identical everywhere. Counties differ in wage levels, labor market structure, income, and education. A wage floor that is very binding in one place may barely matter in another. That is why I think it makes more sense to study the county-level response and not just treat every county in a state as if exposure is identical.

## APIs & Data

This project will use three APIs.

### BEA API
This will be the main source for county-industry wage data. I used BEA data in my original thesis for total wages at the county by industry by year level, and I plan to do the same here. I may also use BEA employment data if needed for the constructed variable. In the original paper, BEA was the source for county-level wage totals in Retail Trade, Accommodation and Food Services, and Other Services.

### Census ACS API
This will be used for county-level demographic and socioeconomic controls. In the original paper, the main ACS controls were median household income, the proportion of the population with a bachelor’s degree, and population age 16 and older. I plan to keep those because they help account for differences in county economic conditions, human capital, and labor supply.

### FRED API
This will be used to collect state minimum wage data over time. This gives me the policy variable and lets me identify when different states began implementing increases.

I already obtained the API access needed, and I will include screenshots of that access and the documentation pages at the end of the proposal.

## Variables

The three main variables for the project will be:

1. County-Industry Total Wages
This is the main outcome variable. It will come from the BEA API and will be collected for the same three industries from the original thesis:

Retail Trade

Accommodation and Food Services

Other Services

2. State Minimum Wage
This will come from the FRED API and will be the main policy variable.

3. Minimum Wage Bite Index
This is the main constructed variable for the project. It will measure how close the state minimum wage is to the average pay in a county-industry-year. The idea is simple: if the minimum wage is already close to what workers in that county and industry are earning on average, then the policy is more likely to bind and have a noticeable effect. If the average local wage is much higher, the minimum wage should matter less.

Minimum Wage Bite Index = State Minimum Wage / Estimated County-Industry Hourly Pay

This variable does not come directly from one API response. It has to be built from the API data, which is why I think it works well for the assignment requirement that at least one variable be calculated and not just downloaded ready-made.

## Main Controls

I also plan to include the same main controls from the original paper:

Median household income

Share of the population with a bachelor’s degree

Population age 16 and older

On top of that, I will use county fixed effects and year fixed effects again. In the original thesis, those were included to account for time-invariant county differences and broader national shocks affecting all counties in a given year.

I will probably also keep some version of the population cutoff from the original study so that very small counties with noisier ACS estimates do not drive the results. In the thesis, the sample was restricted to counties with over 60,000 residents for reliability reasons.

Design Improvements Over the Original Paper

The biggest improvement is that this will not be limited to one treatment year. In the original paper, treated counties were in states that began implementing minimum wage hikes in 2016, while the control group was made up of counties in states without major changes during the study period. The treatment states were California, New York, New Jersey, Illinois, Minnesota, and Massachusetts, while the control states were Texas, Florida, Georgia, Ohio, Pennsylvania, and North Carolina.

That setup was useful for a first pass, but it was also restrictive. It forced everything into one treated versus control comparison and only gave a short pre-treatment window. In this updated version, I want to go farther back in time so I can actually test pre-trends better, and I want to include more states with different implementation years instead of forcing the analysis around 2016 only.

Because of that, I plan to use a Callaway-Sant’Anna style staggered difference-in-differences approach rather than relying only on the original two-way fixed effects event study. That should fit the data better because minimum wage increases are naturally staggered across states and over time.

### So the new version improves the old one by:

using longer pre-trends

allowing for different state implementation years

using a staggered treatment design

adding the Minimum Wage Bite Index as a more direct measure of policy exposure

I still plan to stop the main analysis before the COVID period gets too messy, because once the pandemic starts it becomes much harder to separate wage effects from broader labor market shocks.
